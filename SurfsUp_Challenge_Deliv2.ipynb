```python
# Dependencies
import numpy as np
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
import datetime as dt
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")
engine
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view/check all of the classes that automap found
Base.classes.keys()
# the output is the expected and desired result for the cell
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
#6. Design a query to retrieve the data from June of tobs data 
#Starting from the last data point in the database. 

# Calculate the date one month (dec) {module 9 section 9.2.1 to 9.2.5}

# 3 Perform a query to retrieve the data of tobs
results_2016 = []
results_2016 = session.query(Measurement.date, Measurement.tobs).filter(Measurement.date <= '2016-12-31').filter(Measurement.date >= '2016-12-01' ).all()
print(results_2016)
# reference: https://stackoverflow.com/questions/8895208/sqlalchemy-how-to-filter-date-field
```
